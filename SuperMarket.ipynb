```python
South_df = df[df['Region'] == 'South']
West_df = df[df['Region'] == 'West']
East_df = df[df['Region'] == 'East']
Central_df = df[df['Region'] == 'Central']
North_df = df[df['Region'] == 'North']

```

```python
len(North_df)
```

```python
# orders per region , Note : we have 0 orders from "North" so we We excluded 
region_counts = [len(South_df), len(West_df), len(East_df) , len(Central_df)]
labels = ['South', 'West', 'East' , 'central']

plt.figure(figsize=(8,6))
plt.pie(region_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Orders per region ")
plt.axis('equal')
plt.show()

sns.countplot(data=df, x='Region')
plt.title('Orders per region')
plt.show()


print(f'Number of orders from the South : {len(South_df)}')
print(f'Number of orders from the West : {len(West_df)}')
print(f'Number of orders from the East : {len(East_df)}')
print(f'Number of orders from the Central : {len(Central_df)}')

```

```python
# top cities in each region
Top_West = West_df['City'].value_counts().head()
Top_South = South_df['City'].value_counts().head()
Top_East = East_df['City'].value_counts().head()
Top_Central = Central_df['City'].value_counts().head()

Top_List = [
    ('West', Top_West),
    ('East', Top_East),
    ('South', Top_South),
    ('Central', Top_Central)
]

def ploting(top_list):
    plt.figure(figsize=(12, 8))
    
    for i, (region, data) in enumerate(top_list, start=1):
        ax = plt.subplot(2, 2, i)
        bars = sns.barplot(x=data.values, y=data.index, hue=data.index, palette='Blues_r', legend=False, ax=ax)

        for bar in bars.patches:
            value = int(bar.get_width())
            x = bar.get_x() + bar.get_width() / 2
            y = bar.get_y() + bar.get_height() / 2
            ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, str(value), va='center')

        ax.set_title(f'Top 5 Cities in {region} Region')
        ax.set_xlabel('Number of Orders')
        ax.set_ylabel('City')

    plt.tight_layout()
    plt.show()

ploting(Top_List)



        
```
