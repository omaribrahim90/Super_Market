```python
import os
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub


path = kagglehub.dataset_download("vivek468/superstore-dataset-final")
print("Path to dataset files:", path)


files = os.listdir(path)
print("Files:", files)

file_path = os.path.join(path, 'Sample - Superstore.csv')
df = pd.read_csv(file_path, encoding='latin1') 
print(df.columns)
```

# Data cleaning 

```python

df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Month'] = df['Order Date'].dt.month
```

# Analysis

```python
# orders per region , Note : we have 0 orders from "North" so we We excluded 
region_counts = df['Region'].value_counts()

plt.figure(figsize=(8,6))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Orders per region")
plt.show()

sns.countplot(data=df, x='Region')
plt.title('Orders per region')
plt.show()

region_counts
```

```python
# top cities in each region

Top_List = [
    (region, df[df['Region'] == region]['City'].value_counts().head())
    for region in df['Region'].unique()
]

def plot_top_cities(top_list):
    plt.figure(figsize=(12, 8))
    palettes = ['Blues_r', 'Greens_r', 'Oranges_r', 'Purples_r'] 
    
    for i, (region, data) in enumerate(top_list, start=1):
        ax = plt.subplot(2, 2, i)
        bars = sns.barplot(
            x=data.values,
            y=data.index,
            hue=data.index,
            palette=palettes[i-1],
            legend=False,
            ax=ax
        )

        for bar in bars.patches:
            value = int(bar.get_width())
            text_x = bar.get_width() + (max(data.values) * 0.01) 
            text_y = bar.get_y() + bar.get_height() / 2
            ax.text(text_x, text_y, str(value), va='center')

        ax.set_title(f'Top 5 Cities in {region} Region')
        ax.set_xlabel('Number of Orders')
        ax.set_ylabel('City')

    plt.tight_layout()
    plt.show()

plot_top_cities(Top_List)
```

```python
#profit Per region in percentage 
table = df.pivot_table(index='Region', values=['Sales', 'Profit'], aggfunc='sum')
table = table.sort_values(by='Profit', ascending=False)

table['Profit_Margin'] = (table['Profit'] / table['Sales']) * 100



table['Profit_Margin'].plot(kind='bar')
plt.xlabel('Region')  
plt.ylabel('Profit margin in percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

table
```

```python
# Most purchased products
top_product = df['Product Name'].value_counts().head()

top_product.plot(kind='bar', title='Top 5 Best-Selling Products')
plt.xlabel('Product Name')  
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(5)
top_products.plot(kind= 'barh')
plt.xlabel('Sales')
plt.ylabel('Product name')
plt.show()
```

```python
# best selling products for each region
top_products_per_region = (
    df.groupby(['Region', 'Product Name'])['Sales']
    .sum()
    .reset_index()
    .sort_values(['Region', 'Sales'], ascending=[True, False])
    .groupby('Region')
    .head(3)
)

plt.figure(figsize=(12,6))
sns.barplot(data=top_products_per_region, x='Region', y='Sales', hue='Product Name')
plt.title('Top 3 Best-Selling Products per Region')
plt.ylabel('Total Sales')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

top_products_per_region
```

```python
sns.countplot(data=df , x='Segment')
plt.title('Orders by segment')
plt.show()

df['Segment'].value_counts()

seg_analysis = df.groupby('Segment').agg({
    'Sales':'sum',
    'Profit':'sum',
    'Segment':'count'
}).rename(columns={'Segment':'Orders'})

seg_analysis['AOV'] = seg_analysis['Sales'] / seg_analysis['Orders']
seg_analysis['Profit_per_Order'] = seg_analysis['Profit'] / seg_analysis['Orders']

seg_analysis

```

```python
Seg_product = df.groupby('Segment')['Profit'].sum()
Seg_product.plot(kind='bar' , color='skyblue')
plt.title('Sales by Product Segment')
plt.xlabel('Segment')
plt.ylabel('sales')
plt.xticks(rotation=45)
plt.show()

Seg_product
```

```python
# monthly sales
monthly_orders = df.groupby('Month')['Sales'].sum().reset_index()
monthly_orders.plot(kind='line', x='Month', y='Sales')
plt.title('Monthly Sales')
plt.show()
monthly_orders

monthly_profit= df.groupby('Month')['Profit'].sum().reset_index()
monthly_profit.plot(kind='line', x='Month', y='Profit')
plt.title('Monthly profit')
plt.show()
monthly_orders

```
