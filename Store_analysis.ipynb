```python
import os
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub


path = kagglehub.dataset_download("vivek468/superstore-dataset-final")
print("Path to dataset files:", path)


files = os.listdir(path)
print("Files:", files)

file_path = os.path.join(path, 'Sample - Superstore.csv')
df = pd.read_csv(file_path, encoding='latin1') 
print(df.columns)




```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

EDA

```python
South_df = df[df['Region'] == 'South']
West_df = df[df['Region'] == 'West']
East_df = df[df['Region'] == 'East']
North_df = df[df['Region'] == 'North']

```

```python
len(North_df)
```

```python
# orders per region , Note : we have 0 orders from "North" so we We excluded 
region_counts = [len(South_df), len(West_df), len(East_df)]
labels = ['South', 'West', 'East']

plt.figure(figsize=(8,6))
plt.pie(region_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Number of orders per region ")
plt.axis('equal')
plt.show()

print(f'Number of orders from the South : {len(South_df)}')
print(f'Number of orders from the West : {len(West_df)}')
print(f'Number of orders from the East : {len(East_df)}')
```
